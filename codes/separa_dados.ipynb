```python
""" importa as bibliotecas necessárias """
import pandas as pd
```

```python
""" essa célula abre a base de dados """

# abre o arquivo
path_file = '../data/dados_agrupados.csv'

file = pd.read_csv(path_file)

# print(file.head(1))
```

```python
""" essa célula identifica os códigos de diagnóstico presentes na base de dados """

# itera sobre o arquivo
for index, row in file.iterrows():
    # seleciona a coluna 'diagnosis' e separa os codigos
    diagnosis_data = row['diagnosis']
    diagnosis_list = diagnosis_data.split(',')
    
    # abre o arquivo de codigos, le e checa se o codigo ja esta no arquivo. se nao tiver, escreve
    try: 
        codes_file = open('../data/codigos_arritmias.txt', 'r')
        data = codes_file.read()
        codes_file.close()
        for i in range(len(diagnosis_list)):
            if diagnosis_list[i] in data:
                print("código já lido")
            else:
                codes_file = open('../data/codigos_arritmias.txt', 'a')
                codes_file.write(f'{diagnosis_list[i]}\n')
                codes_file.close()
    except FileNotFoundError:
        print("Não foi possível abrir o arquivo.")
    
```

```python
""" essa célula é responsável por separar a base de dados em 2 arquivos:
arquivo_1: pacientes sem alteração no exame
arquivo_2: pacientes com alteração no exame """

# colunas do DataFrame
columns = ['path','filename','age','sex','diagnosis','leads','frequency','samples','lead_I_first',
        'lead_I_last','lead_II_first','lead_II_last','lead_III_first','lead_III_last','lead_aVF_first',
        'lead_aVF_last','lead_aVL_first','lead_aVL_last','lead_aVR_first','lead_aVR_last','lead_V1_first',
        'lead_V1_last','lead_V2_first','lead_V2_last','lead_V3_first','lead_V3_last','lead_V4_first',
        'lead_V4_last','lead_V5_first','lead_V5_last','lead_V6_first','lead_V6_last']

# armazena o caminho dos arquivos
path_normal = '../data/exames_sem_alteracao.csv'
path_abnormal = '../data/exames_com_alteracao.csv'

# lista de dicionários para armazenar as linhas
normal_dict = []
abnormal_dict = []

# le o arquivo original e diferencia os pacientes com anormalidade ou nao
for index, row in file.iterrows():
    diagnosis_data = row['diagnosis']
    diagnosis_list = diagnosis_data.split(',')

    line = row.to_dict()

    # codigo de ritmo sinusal = 426783006
    if len(diagnosis_list) == 1 and diagnosis_list[0] == '426783006':
        normal_dict.append(line)
    else:
        abnormal_dict.append(line)

# criação dos datadrames a partir das listas de dicionários
if normal_dict:
    normal = pd.DataFrame(normal_dict, columns=columns)
    normal.to_csv(path_normal, mode='a', index=False)

if abnormal_dict:
    abnormal = pd.DataFrame(abnormal_dict, columns=columns)
    abnormal.to_csv(path_abnormal, mode='a', index=False)
```
