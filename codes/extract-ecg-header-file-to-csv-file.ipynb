```python

import pandas as pd
import os
import glob
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
ROOT_PATH = 'dataset_chapmanShaoxing'
```

```python
path = ROOT_PATH+ '/*.hea'
```

```python
data_list =[]

for filepath in glob.glob(path, recursive=True):
    #print(filepath)
    with open(os.path.join(filepath), 'r') as f:
        lines = f.readlines()
        
        # Subject Id
        filename = lines[0].split()[0]

        # Total lead Number
        leads = lines[0].split()[1]

        # Sampling Frequency
        samp_freq = lines[0].split()[2]

        # Total Samples
        samples = lines[0].split()[3]

        # Lead I data
        first_value_I = lines[1].split()[5]
        last_value_I = lines[1].split()[6]
        lead_I = lines[1].split()[7]

        # Lead II data
        first_value_II = lines[2].split()[5]
        last_value_II = lines[2].split()[6]
        lead_II = lines[2].split()[7]


        # Lead III data
        first_value_III = lines[3].split()[5]
        last_value_III = lines[3].split()[6]
        lead_III = lines[3].split()[7]


        # Lead aVR data
        first_value_aVR = lines[4].split()[5]
        last_value_aVR = lines[4].split()[6]
        lead_aVR = lines[4].split()[7]


        # Lead aVL data
        first_value_aVL = lines[5].split()[5]
        last_value_aVL = lines[5].split()[6]
        lead_aVL = lines[5].split()[7]



        # Lead aVF data
        first_value_aVF = lines[6].split()[5]
        last_value_aVF = lines[6].split()[6]
        lead_aVF = lines[6].split()[7]


        # Lead V1 data
        first_value_v1= lines[7].split()[5]
        last_value_v1 = lines[7].split()[6]
        lead_V1 = lines[7].split()[7]

        # Lead V2 data
        first_value_v2 = lines[8].split()[5]
        last_value_v2 = lines[8].split()[6]
        lead_V2 = lines[8].split()[7]


        # Lead V3 data
        first_value_v3 = lines[9].split()[5]
        last_value_v3 = lines[9].split()[6]
        lead_V3 = lines[9].split()[7]

        # Lead V4 data
        first_value_v4 = lines[10].split()[5]
        last_value_v4 = lines[10].split()[6]
        lead_V4 = lines[10].split()[7]


        # Lead V5 data
        first_value_v5 = lines[11].split()[5]
        last_value_v5 = lines[11].split()[6]
        lead_V5 = lines[11].split()[7]


        # Lead V6 data
        first_value_v6 = lines[12].split()[5]
        last_value_v6 = lines[12].split()[6]
        lead_V6 = lines[12].split()[7]



        # Age
        age = lines[13].split()[1]

        # Sex
        sex = lines[14].split()[1]

        # Diagnosis
        dx = lines[15].split()[1]
        
        
        #print(filename, age, sex, dx, rx, hx)

        header_df = pd.DataFrame({
            
            "path":[os.path.splitext(filepath)[0]],
            "filename":[filename],
            "age":[age],
            "sex":[sex],
            "diagnosis":[dx],
            "leads":[leads],
            "frequency": [samp_freq],
            "samples":[samples],
            "lead_I_first": [first_value_I],
            "lead_I_last":[last_value_I],
            "lead_II_first": [first_value_II],
            "lead_II_last":[last_value_II],
            "lead_III_first": [first_value_III],
            "lead_III_last":[last_value_III],
            "lead_aVF_first": [first_value_aVF],
            "lead_aVF_last":[last_value_aVF],
            "lead_aVL_first": [first_value_aVL],
            "lead_aVL_last":[last_value_aVL],
            "lead_aVR_first": [first_value_aVR],
            "lead_aVR_last":[last_value_aVR],
            "lead_V1_first": [first_value_v1],
            "lead_V1_last":[last_value_v1],
            "lead_V2_first": [first_value_v2],
            "lead_V2_last":[last_value_v2],
            "lead_V3_first": [first_value_v3],
            "lead_V3_last":[last_value_v3],
            "lead_V4_first": [first_value_v4],
            "lead_V4_last":[last_value_v4],
            "lead_V5_first": [first_value_v5],
            "lead_V5_last":[last_value_v5],
            "lead_V6_first": [first_value_v6],
            "lead_V6_last":[last_value_v6],


        
        })
        data_list.append(header_df)

```

```python
df = pd.concat(data_list).sort_values(by='path')

df.drop_duplicates(inplace=True)

df.to_csv(  './dados_agrupados.csv', index=False)
```
